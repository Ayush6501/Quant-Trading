# Momentum

Momentum is the measure of the rate of acceleration of a stock's price. The main way to measure momentum is through dividing a stock's current price by the stock's price a given number of days ago.

$$
momentum = {p_i \over {p_{i - n}}}
$$

```python
# define time period before stock can be traded
gap = 14
CocaCola['Momentum'] = CocaCola['PX_LAST'].pct_change(periods=gap)
CocaCola.head()
```

## Strategy Implementation - Simple BackTesting

```python
KO_or_PEP = [0]*14

for i in range(len(CocaCola)):
    if CocaCola['Momentum'][i] > Pepsi['Momentum'][i]:
        KO_or_PEP.append("PEP")
    else:
        KO_or_PEP.append("KO")

print(KO_or_PEP)
```

```python
KO_price = CocaCola.iloc[:,1]
PEP_price = Pepsi.iloc[:,1]

init_value = 1000
value = init_value
last_move_day = 0
position = ""

days = range(gap, len(CocaCola['Momentum']))
values = []

for i in range(gap, len(CocaCola['Momentum'])):
    if i == gap:
        position = KO_or_PEP[i]
    else:
        if KO_or_PEP[i - gap] != KO_or_PEP[last_move_day - gap]:
            if position == "PEP":
                value *= PEP_price[i] / PEP_price[last_move_day]
                position = "KO"
            else:
                value *= KO_price[i] / KO_price[last_move_day]
                position = "PEP"
            last_move_day = i
            value -= 0.675 # add a fee!
    values.append(value)

print(value - init_value, "percent gain!")

plt.plot(days, values, label="Our Position")
plt.xlabel("Dates")
plt.ylabel("Money (benchmarked)")
plt.title("KO Momentum over time (n = 7 days)")
plt.legend()
plt.show()
```

## Advanced BackTesting

Now we will allocate funds to PEP and KO proportionally against their momentum instead of absolutely (hint: this means the array of positions we'll store and add to will be numeric!)

Our decision-making process will be as follows:


* if PEP and KO have momentum above 1, allocate $\frac{m_{PEP} - 1}{m_{PEP} + m_{KO} - 2}$ funds to PEP and $\frac{m_{KO} - 1}{m_{PEP} + m_{KO} - 2}$ to KO.
* if only PEP has momentum above 1, allocate all funds to PEP.
* if only KO has momentum above 1, allocate all funds to KO.
* if neither KO nor PEP have momentum above 1, get out of both positions.

We encourage you to change this process and implement your own version once you're done to best experiment with this strategy!


```python
print(Pepsi_momentum_list)
print(Coke_momentum_list)
```

```python
KO_positions = []
PEP_positions = []

for i in range(gap, len(Pepsi_momentum_list)):
    if Pepsi_momentum_list[i] > 0:
        if Coke_momentum_list[i] > 0:
            KO_positions.append((Coke_momentum_list[i] - 1) / (Coke_momentum_list[i] + Pepsi_momentum_list[i] - 2))
            PEP_positions.append((Pepsi_momentum_list[i] - 1) / (Coke_momentum_list[i] + Pepsi_momentum_list[i] - 2))
        else:
            KO_positions.append(0)
            PEP_positions.append(1)
    else:
        if Coke_momentum_list[i] > 0:
            KO_positions.append(1)
            PEP_positions.append(0)
        else:
            KO_positions.append(0)
            PEP_positions.append(0)

print(KO_positions)
print(PEP_positions)
```

```python
KO_price = CocaCola.iloc[:,1]
PEP_price = Pepsi.iloc[:,1]

init_value = 10000
value_out = init_value
value_in = 0
last_move_day = 0
KO_position = 0
PEP_position = 0

days = range(gap, len(Pepsi_momentum_list))
values = []

for i in range(gap, len(Pepsi_momentum_list)):
    # what we had in PEP and KO
    PEP_init_holdings = value_in * PEP_position
    KO_init_holdings = value_in * KO_position
    # what we have now in PEP and KO:
    PEP_now_holdings = 0
    KO_now_holdings = 0
    if i > gap:
        PEP_now_holdings = PEP_init_holdings * PEP_price[i - gap] / PEP_price[i - gap - 1] # PEP_init_holdings * now price / yesterday price
        KO_now_holdings = KO_init_holdings * KO_price[i - gap] / KO_price[i - gap - 1] # KO_init_holdings * now price / yesterday price
    PEP_position = PEP_positions[i - gap]
    KO_position = KO_positions[i - gap]

    total_val = value_out + value_in + PEP_now_holdings - PEP_init_holdings + KO_now_holdings - KO_init_holdings
    # subtract cost for each trade (commission)
    total_val -= 0.675 # add a fee!
    value_out = total_val * (1 - PEP_position - KO_position)
    value_in = total_val * (PEP_position + KO_position)

    values.append(value_in + value_out)

print((value_in + value_out - init_value)/100, "percent gain!")

plt.plot(days, values, label="Our Position")
plt.xlabel("Dates")
plt.ylabel("Money (benchmarked)")
plt.title("KO Momentum over time (n = 7 days)")
plt.legend()
plt.show()
```
